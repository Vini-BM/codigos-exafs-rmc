# Separação dos blocos contidos nos arquivos 'restart*'

## Ideias gerais

## Programa

```python
#Abrir arquivo
arq = 'restart600' #Escolha arbitrária, apenas para teste
f = open(arq)
content = f.readlines()
```

```python
#Encontrar início e fim
line_num = 0
for line in content:
    line_num += 1
    if 'ATOMS' in line:
        inicio = line_num-1 #Encontrar número da linha onde os blocos iniciam
    if 'SPECTRA' in line:
        fim = line_num-1 #Encontrar número da linha onde os blocos terminam
        break
print(inicio) #Para teste
print(fim) #Para teste

#Não foi preciso usar a variável 'fim', mas é bom tê-la salva para testes
```

```python
#Encontrar o número de átomos N
N_line = inicio+1 #Primeira linha em que N aparece
N = int(content[N_line]) #Número de átomos
num_linhas = N+2 #Número de linhas por bloco
print('Número de átomos:',N)
```

```python
#Separar blocos
num_blocos = int((fim - N_line)/(num_linhas)) #Número de blocos (assumindo simetria no arquivo)
#print(num_blocos)
a = N_line #Contador -> evitar percorrer todo o arquivo
blocos = {} #Dicionário com os blocos, apenas para controle
for i in range(0, num_blocos): #Loop maior: criar identificação para os blocos
    nome_bloco = 'dados_'+str(i) #String com nome do bloco
    locals()[nome_bloco] = [] #Criar lista chamada nome_bloco; a cada iteração o nome é atualizado e uma nova lista é criada
    lista = locals()[nome_bloco]
    nome_arquivo = arq + '-' + nome_bloco #Nome do arquivo a ser criado: um arquivo por bloco
    f = open(nome_arquivo, 'w')
    for j in range(a, a+num_linhas): #Loop menor: ler dados das N+2 linhas que vão em cada bloco
        texto_linha = content[j].strip().split()
        #print(texto_linha)
        lista.append(texto_linha) #Adiciona o conteúdo da linha na lista
        for line in texto_linha:
            f.write(line + ' ') #Escrever conteúdo da linha (lista) no arquivo
        f.write('\n')
        j=+1       
    a=a+num_linhas #Aumento do contador para o loop maior
    blocos['Bloco ' + str(i)] = nome_bloco #Adicionar nome do bloco no dicionário
    i=+1

#Testes:
#print(a)
#print(fim)
#dados_0
#dados_1
```
